```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from datetime import datetime
```

```python
def just_time(timestamp):
    dates = []
    for stamp in timestamp:
        stamp = datetime.strptime(stamp, '%Y-%m-%d %H:%M:%S')
        stamp = stamp.strftime('%-d.%-m\n%H:%M')
        dates.append(stamp)
    return dates
```

```python
def setup(ax, title):
    """Set up common parameters for the Axes in the example."""
    # only show the bottom spine
    ax.yaxis.set_major_locator(ticker.NullLocator())
    ax.spines.right.set_color('none')
    ax.spines.left.set_color('none')
    ax.spines.top.set_color('none')

    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(which='major', width=1.00, length=5)
    ax.tick_params(which='minor', width=0.75, length=2.5)
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 1)
    ax.text(0.0, 0.2, title, transform=ax.transAxes,
            fontsize=14, fontname='Monospace', color='tab:blue')
```

```python
df1 = pd.read_csv('working_data.csv')[1000:1500]
timestamp = df1['Timestamp']
smp_real = np.array(df1['SMP'])

df2 = pd.read_csv('working_data_peak.csv')[1000:1500]
smp_peak = np.array(df2['SMP'])
gen = np.array(df1['Energy_Generation'])/100 + 0.07
timestamp = just_time(timestamp)


fig, ax = plt.subplots(1, 1, figsize=(24, 6))
# Null Locator
#setup(axs, title="NullLocator()")
ax.xaxis.set_major_locator(ticker.MultipleLocator(24))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(24))
plt.setp(ax.xaxis.get_majorticklabels(), rotation = 0)
ax.plot(timestamp, smp_real, c='lightgreen', label='Real-Time Pricing (RTP)')
ax.plot(timestamp, smp_peak, c='magenta', label='Time-Of-Use (TOU)')
ax.plot(timestamp, gen, c='orange', alpha=0.3)
ax.set_ylim(0.05, 0.1)
ax.grid(linestyle='-', linewidth=1)
plt.legend()
#plt.margins(x=0, y=0)
plt.savefig('pricing.png', bbox_inches='tight')
plt.show()


new_df = pd.DataFrame(list(zip(timestamp, smp_real, smp_peak)), columns=['Tmestamp', 'Real time pricing', 'Time-of-Use'])
print(new_df)
new_df.to_csv('pricing.csv')
```

```python
df1 = pd.read_csv('working_data.csv')
df2 = pd.read_csv('working_data_peak.csv')

s_high_reals = []
s_high_peaks = []
step_1 = None
s_gen_high = 0
for i in range(200,300):
    two_days = 4*24*2
    step_0 = i*two_days
    step_1 = (i+1)*two_days
    smp_real = np.array(df1['SMP'])[step_0:step_1]
    smp_peak = np.array(df2['SMP'])[step_0:step_1]
    gen = np.array(df1['Energy_Generation'])[step_0:step_1]


    median_smp_real = np.mean(smp_real)
    median_smp_peak = np.mean(smp_peak)
    median_gen = np.median(gen)
    #print(median_smp_real, median_smp_peak, median_gen)
    s_high_real = 0
    s_high_peak = 0
    s_gen_high = 1
    for i in range(len(gen)):
        smp_real_high = smp_real[i] > median_smp_real
        smp_peak_high = smp_peak[i] > median_smp_peak
        gen_high = gen[i] > 0
        if smp_real_high and gen_high: s_high_real += 1
        if smp_peak_high and gen_high: s_high_peak += 1
        if gen_high: s_gen_high += 1

    s_high_reals.append(s_high_real/s_gen_high)
    s_high_peaks.append(s_high_peak/s_gen_high)

print(step_1)
print(np.mean(s_high_reals), np.mean(s_high_peaks))
```
